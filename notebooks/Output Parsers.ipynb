```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.messages import HumanMessage
from langchain_core.output_parsers import StrOutputParser
```

```python
chat = ChatOpenAI(model_name = 'gpt-4', seed = 365, temperature = 0, max_tokens = 100)
```

```python
message_human = HumanMessage(content = "Can you give me an interesting fact I probably didn't know about?")
```

```python
response = chat.invoke([message_human])
```

```python
response
```

```python
str_output_parser = StrOutputParser()
```

```python
response_parsed = str_output_parser.invoke(response)
```
