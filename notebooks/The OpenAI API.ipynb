```python
import os
import openai
```

```python
openai.api_key = os.getenv('OPENAI_API_KEY')
```

```python
client = openai.OpenAI()
```

```python
completion = client.chat.completions.create(model = 'gpt-4', messages = [{'role':'system',
                                                                          'content':'''You are Marv, a chatbot that classifies tweets as 
                                                                          positive negative or neutral. '''},
                                                                         {'role':'user', 
                                                                          'content':''' Such a great idea, lets implement this! '''}])
```

```python
completion
```
